```python
# Variável que vai armazenar uma lista, onde serão guardados as listas com os dados de cada página, com cada repositório
repos_list = []

# For loop que percorre cada página (de 1 a 5, não chega no 6)
for page_num in range(1, 6):
    # Tenta
    try:
        # Cria a variável com a URL determinando a página
        url_page = f'{final_url}?page={page_num}'
        # Faz a requisição com a URL com a página
        response = requests.get(url_page, headers=headers)
        # Insere a resposta na variável repos_list
        repos_list.append(response.json())
    except:
        # Se não der, append None
        repos_list.append(None) 
```

```python
len(repos_list)
# Retorna 5, pois são 5 páginas
```

```python
len(repos_list[0])
# Retorna 30, pois são 30 repositórios
```

A lista de dados dos repositórios está *grande demais*.
Como o objetivo é ver a linguagem de programação mais utilizada por cada empresa, vamos manter *apenas o nome dos repos. e as linguagens de programação de cada um*.

```python
repos_list[0][2]['name']
```

O que vamos fazer, é percorrer toda nossa lista, extraindo os nomes, e salvando-os em uma nova lista

```python
repos_name = []

for page in repos_list:
    for repo in page:
        repos_name.append(repo['name'])
```

```python
len(repos_name)
```

```python
repos_lang = []

for page in repos_list:
    for repo in page:
        repos_lang.append(repo['language'])
```

```python
repos_lang[:10]
```

```python
len(repos_lang)
```
